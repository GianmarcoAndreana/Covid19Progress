```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta, date
import plotly.express as px
```

```python
dataCases = pd.read_csv("--path--/Desktop/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
dataCases = dataCases.loc[dataCases["Province/State"].isna(),].iloc[:,1:]
dataDeaths = pd.read_csv("C:/Users/giamm/Desktop/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
dataDeaths = dataDeaths.loc[dataDeaths["Province/State"].isna(),].iloc[:,1:]
dataRecovered = pd.read_csv("C:/Users/giamm/Desktop/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
dataRecovered = dataRecovered.loc[dataRecovered["Province/State"].isna(),].iloc[:,1:]
```

```python
def sevenDaysCases(country):
    days = pd.date_range(start="2020-01-22",end=(datetime.today()-timedelta(days=1)))
    cases = dataCases.loc[dataCases["Country/Region"] == country,:].iloc[:,3:].values.tolist()[0]
    ma = cases[0:6]
    count = 0
    for i in cases[6:]:
        ma.append(i-cases[count])
        count += 1
    fig = px.line(x=days, y=ma)
    fig.update_layout(
    title="Covid cases in "+country+", weekly increase",
    xaxis_title="Time",
    yaxis_title="Cases")
    fig.show()
    
def sevenDaysDeaths(country):
    days = pd.date_range(start="2020-01-22",end=(datetime.today()-timedelta(days=1)))
    deaths = dataDeaths.loc[dataDeaths["Country/Region"] == country,:].iloc[:,3:].values.tolist()[0]
    ma = deaths[0:6]
    count = 0
    for i in deaths[6:]:
        ma.append(i-deaths[count])
        count += 1
    fig = px.line(x=days, y=ma)
    fig.update_layout(
    title="Covid deaths in "+country+", weekly increase",
    xaxis_title="Time",
    yaxis_title="Deaths")
    fig.show()
    
def sevenDaysRecovered(country):
    days = pd.date_range(start="2020-01-22",end=(date(2021,8,5)-timedelta(days=1)))
    recovered = dataRecovered.loc[dataRecovered["Country/Region"] == country,:].iloc[:,3:564].values.tolist()[0]
    ma = recovered[0:6]
    count = 0
    for i in recovered[6:]:
        ma.append(i-recovered[count])
        count += 1
    fig = px.line(x=days, y=ma)
    fig.update_layout(
    title="Covid recovered in "+country+", weekly increase",
    xaxis_title="Time",
    yaxis_title="Individuals")
    fig.show()
    
def dailyCases(country):
    days = pd.date_range(start="2020-01-22",end=(datetime.today()-timedelta(days=1)))
    cases = dataCases.loc[dataCases["Country/Region"] == country,:].iloc[:,3:].values.tolist()[0]
    ma = [cases[0]]
    count = 0
    for i in cases[1:]:
        ma.append(i-cases[count])
        count += 1
    fig = px.line(x=days, y=ma)
    fig.update_layout(
    title="Covid cases in "+country+", daily increase",
    xaxis_title="Time",
    yaxis_title="Cases")
    fig.show()
    
def dailyDeaths(country):
    days = pd.date_range(start="2020-01-22",end=(datetime.today()-timedelta(days=1)))
    deaths = dataDeaths.loc[dataDeaths["Country/Region"] == country,:].iloc[:,3:].values.tolist()[0]
    ma = [deaths[0]]
    count = 0
    for i in deaths[1:]:
        ma.append(i-deaths[count])
        count += 1
    fig = px.line(x=days, y=ma)
    fig.update_layout(
    title="Covid deaths in "+country+", daily increase",
    xaxis_title="Time",
    yaxis_title="Deaths")
    fig.show()
    
def dailyRecovered(country):
    days = pd.date_range(start="2020-01-22",end=(date(2021,8,5)-timedelta(days=1)))
    recovered = dataRecovered.loc[dataRecovered["Country/Region"] == country,:].iloc[:,3:564].values.tolist()[0]
    ma = [recovered[0]]
    count = 0
    for i in recovered[1:]:
        ma.append(i-recovered[count])
        count += 1
    fig = px.line(x=days, y=ma)
    fig.update_layout(
    title="Covid recovered in "+country+", daily increase",
    xaxis_title="Time",
    yaxis_title="Individuals")
    fig.show()
```
